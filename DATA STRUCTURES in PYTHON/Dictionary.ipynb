# Working on Dictionary (data structure) in Python
Several operations involved in the ***dictionary*** are:

**1. Adding New Records**

**2. Viewing Records**

**3. Searching Records**

**4. & Removing data**

```python
d1=dict()   #dictionary declared globally

def add_data():
    l=[]
    roll_no=int(input("Enter your Roll No. "))
    name=input("Enter your name ")
    num=int(input("Enter number of subjects "))                        #input from the user
    for i in range(1,num+1):
        print("Enter Subject",i)
        subject=input()
        l.append(subject)
    d1[roll_no]=[name,l]                                              #adding content to the "dictionary"
    print("Data added successfully in dictionary d1!")
    
        

def view_data():                                                     #viewing records
    flag=0
    d=input("\nDo you want to view: \n1. All data \n\tor \n2. A specific record?\n ")   #asking user if all records or a particular record to be displayed
    
    if(d=='1'):
        if(d1=={}):
            print("\nDictionary is empty!")
        else:
            for i in d1:
                print()
                print(i,end='\t')
                for j in d1[i]:
                    print(j, end='\t')
    elif(d=='2'):
        c=int(input("\nEnter roll no. to view a particular record "))
        for i in d1:
            if(i==c):
                flag=1
                print(i,end='\t')
                for j in d1[i]:
                    print(j,end='\t')
      

        if(flag==0):
            print("\n\t\tNot found")
    else:
        print("\nWrong choice entered")
            
                
def search_data(roll):          #searching records on the basis of "key" i.e. Roll Number in the dictionary
    flag=0
    for i in d1:
        if(i==roll):
            flag=1
            print(i,end='\t')
            for j in d1[i]:
                print(j,end='\t')
    if(flag==0):
        print("\nEntered Roll No. is wrong, try again!!")
        
    
def remove_data(roll):        #removing records on the basis of "key" in the dictionary
    flag=0
    try:
        for i in d1:
            if(i==roll):
                flag=1
                d1.pop(i)
                print("Roll No. ",i,"is removed from the records")
        if(flag==0):
            print("\nSorry not found!")
    except:
        pass
    

ch=''
while(ch.lower()!='n'):
    print("\n1. Add New Data\n2. View Data\n3. Search Records\n4. Remove Data ")  #menu driven program & functions being called
    c1=eval(input("\nWhich operation do you want to perform in a dictionary?"))
    
    if(c1==1):
        add_data()
    elif(c1==2):
        view_data()
    elif(c1==3):
        e1=int(input("\nEnter Roll No, whose record you want to search "))
        search_data(e1)
    elif(c1==4):
        e1=int(input("\nEnter Roll No, whose details you want to delete "))
        remove_data(e1)
    else:
        print("wrong choice!")
    
    ch=input("\nDo you want to continue? y/n ")
print("\n\t\t\t\t\tThank you for exploring \'dictionary\' data structure!")
```
